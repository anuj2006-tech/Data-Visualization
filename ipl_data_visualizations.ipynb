To know that who won most matchs

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y= match_wins.index , x = match_wins.values, palette = 'viridis') # we shuffle here x & y so it will visible clearly
plt.title("Most match win by teams")
```

For toss decisions

```python
sns.countplot( x = df['toss_decision'],palette = 'rainbow')
plt.title('Toss decision trends')
```

Toss winner vs Match winner

```python
toss_match_wins = df[df['toss_winner'] == df['match_winner']]
toss_match_wins_count = toss_match_wins['toss_winner'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=toss_match_wins_count.index, y=toss_match_wins_count.values, palette='coolwarm')
plt.title('Teams Winning Toss and Match')
plt.xlabel('Team')
plt.ylabel('Number of Times Toss & Match Won')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
total_matches = len(df)
toss_match_win_percentage = (len(toss_match_wins) / total_matches) * 100
print(f"Percentage of times toss winning team won the match: {toss_match_win_percentage:.2f}%")
```

10 best bowling figure

```python

df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind = 'barh')
```
